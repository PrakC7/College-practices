```python
import matplotlib.pyplot as plt
import numpy as np

# Make sure x is a NumPy array for element-wise operations
x = np.array([0, 1, 2, 3, 4, 5])

# Create a figure
fig = plt.figure()

# Add axes (the list [left, bottom, width, height] in figure coordinates)
ax = fig.add_axes([0, 0, 1, 1])

# Plot the data
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

axes[0].plot(x, x**2, lw=2)
axes[0].grid(True)

axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', lw=0.5)
```

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 2))

# Set spine colors and linewidth
ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

# Turn off the right spine
ax.spines['right'].set_color('none')  # or ax.spines['right'].set_visible(False)

# Only show ticks on the left y-axis
ax.yaxis.tick_left()

# Optional: add something to plot
ax.plot([0, 1, 2], [0, 1, 0], color='black')

plt.show()
```
